```
import numpy as np
from tensorflow.keras import models,layers,datasets
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```
(x_train,y_train),(x_test,y_test) = datasets.cifar10.load_data()
```

```

x_train,x_test=x_train/255.0,x_test/255.0
print(x_train)
```

```
target_name=['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
print(x_train.shape)
```

```
plt.figure(figsize=(8,8))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i])
    plt.xlabel(target_name[y_train[i][0]])

plt.show()
```

```
plt.figure(figsize=(8,8))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test[i])
    plt.xlabel(target_name[y_test[i][0]])

plt.show()
```

```
#model with dropout :
#model with dropout :
#model with dropout :
#model with dropout :
#model with dropout :
#model with dropout :
#model with dropout :
#model with dropout :
#model with dropout :
```

```
model=models.Sequential()
model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)))
model.add(layers.MaxPooling2D(pool_size=(2,2)))
model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2,2)))
model.add(layers.Dropout(0.35))
model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(10,activation='softmax'))
```

```
model.summary()
```

```
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```
history=model.fit(x_train,y_train,epochs=22,batch_size=19,validation_data=(x_test,y_test),validation_split=0.2,verbose=1)
```

```
test_loss, test_acc = model.evaluate(x_test,y_test,verbose=2)

print(test_acc)
```

```
plt.figure()
plt.plot(history.history['loss'],label='train_loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend(loc='upper left')
plt.title("Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid()
plt.show()
```

```
plt.figure()
plt.plot(history.history['accuracy'], label='train_accurcay')
plt.plot(history.history['val_accuracy'], label='val_accurcy')
plt.legend(loc='upper left')
plt.title("Loss")
plt.xlabel("Epochs")
plt.ylabel("accuracy")
plt.grid()
plt.show()
```

```

print(x_test.shape)
pred=model.predict(x_test)
list_pred=np.argmax(pred,axis=1)
nb_erreur=0
plt.figure(figsize=(4,4))
for i in range(100):
    # plt.subplot(2,2,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    if(target_name[list_pred[i]]!=target_name[y_test[i][0]]):
      nb_erreur+=1
      plt.imshow(x_test[i])
      plt.xlabel("pred_Name:{} \n true_Name:{}".format(target_name[list_pred[i]],target_name[y_test[i][0]]))
      plt.show()
```

```
print("procentage derreur : {}%".format(nb_erreur))
```

```
modification=ImageDataGenerator(
    rotation_range=15,
    zoom_range=0.15,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
)
modification.fit(x_train)
print("le nauveau shape :",x_train.shape)
```

```
exemple=(x_train[0]*255).reshape(1,32,32,3)
img_augmenter=modification.flow(exemple,batch_size=1)
plt.figure(figsize=(8,8))
for i in range(15):
    img=next(img_augmenter)[0]
    plt.subplot(5,3,i+1)
    plt.imshow((img).astype("uint8"))
    plt.axis('off')
plt.suptitle("image augmenter")
plt.show()
```

```
model_optimal=models.Sequential()
model_optimal.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)))
model_optimal.add(layers.MaxPooling2D(pool_size=(2,2)))
model_optimal.add(layers.Conv2D(64,(3,3),activation='relu'))
model_optimal.add(layers.MaxPooling2D(pool_size=(2,2)))
# model_optimal.add(layers.Dropout(0.35))
model_optimal.add(layers.Conv2D(64,(3,3),activation='relu'))
model_optimal.add(layers.MaxPooling2D(pool_size=(2,2)))
model_optimal.add(layers.Flatten())
model_optimal.add(layers.Dense(64,activation='relu'))
model_optimal.add(layers.Dropout(0.5))
model_optimal.add(layers.Dense(10,activation='softmax'))
```

```
model_optimal.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```
history=model_optimal.fit(modification.flow(x_train,y_train,batch_size=20),epochs=50,validation_data=(x_test,y_test),verbose=1)
```

```
test_loss, test_acc = model_optimal.evaluate(x_test,y_test,verbose=2)
print(test_acc)
```
